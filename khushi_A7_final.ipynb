```python
import string
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
document = """
Text analytics is the process of analyzing textual data.
It involves tokenization, stemming, lemmatization, and removing stop words.
Text analytics helps in extracting meaningful insights from text data.
"""
```

```python
sentences = sent_tokenize(document)
print("Sentences:")
print(sentences)
```

```python
words = word_tokenize(document)
print("Words:")
print(words)
```

```python
pos_tags = pos_tag(words)
print("POS Tags:")
print(pos_tags)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words 
                  if word.lower() not in stop_words 
                  and word not in string.punctuation]

print("After Stopword Removal:")
print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("Stemmed Words:")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Lemmatized Words:")
print(lemmatized_words)
```

```python
from collections import Counter

tf = Counter(filtered_words)
print("Term Frequency:")
print(tf)
```

```python
total_terms = len(filtered_words)

tf_normalized = {word: count/total_terms for word, count in tf.items()}
print("Normalized TF:")
print(tf_normalized)
```

```python
documents = [
    "Text analytics is the process of analyzing text data.",
    "Tokenization and stemming are steps in text analytics.",
    "Text data requires preprocessing and cleaning."
]
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(documents)

print("Feature Names:")
print(vectorizer.get_feature_names_out())

print("TF-IDF Matrix:")
print(tfidf_matrix.toarray())
```
