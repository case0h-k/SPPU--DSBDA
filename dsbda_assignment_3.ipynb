```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Age_Group': ['Young', 'Young', 'Adult', 'Adult', 'Senior', 'Senior'],
    'Income': [20000, 25000, 40000, 42000, 30000, 28000]
}

df = pd.DataFrame(data)
print(df)
```

```python
grouped_stats = df.groupby('Age_Group')['Income'].agg(['mean', 'median', 'min', 'max', 'std'])
print(grouped_stats)
```

```python
income_list = df.groupby('Age_Group')['Income'].apply(list)
print(income_list)
```
