```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic["survived_label"] = titanic["survived"].map({0: "Not Survived", 1: "Survived"})

plt.figure(figsize=(8,6))

sns.boxplot(
    x="sex",
    y="age",
    hue="survived_label",   # use the new column
    data=titanic,
    palette="Set1"
)

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

# Legend will automatically show the correct labels in top right
plt.legend(title="Survival Status", loc="upper right")

plt.show()
```
